##### Create new column and new row to calculate the average of each person and each brand.

```python
# Add one row to calculate the average of each brand in the last row.
for i in df.columns:
    df.loc[-1, i] = df.loc[0:-1, i].mean()
df.loc[-1, "Critics"] = "avg_brand"
```

```python
# Add one column to calculate the average of each person in the last column.
df["avg_person"] = df.iloc[0:-1, 1:].mean(axis=1)
```

##### Check the Data again, we have add brand average and person average now.

```python
df
```

##### Do some pre-work to calculate the mean of the whole data.

```python
df1 = df.drop(["Critics"], axis=1)
df1 = df1.drop(["avg_person"], axis=1)
df1 = df1.drop(labels=-1, axis=0)
```

```python
df.iloc[-1, -1] = df1.stack().dropna().mean()
```

```python
df
```

```python
df_avg = df.copy()
```

* We got the average of whole data in the right down corner cell now.

##### Now, let's add the new column and new row to indicate the difference between the avg_person(avg_brand) and the average of the whole data.

```python
# Add new column to show the difference between avg_person and average of the full data.
df["person_diff"] = df.iloc[:-1]["avg_person"] - df.iloc[-1, -1]
```

```python
# Add new row to show the difference between avg_brand and average of the full data.
for i in df.columns[1:9]:
    df.loc[-2, i] = df.loc[-1, i] - df_avg.iloc[-1, -1]
df.loc[-2, "Critics"] = "brand_diff"
```

##### Fill out missing value with Global Baseline Estimate algorithm.

```python
df.fillna(False, inplace=True)
```

```python
for index in df.index:
    for column in df.columns[1:]:
        if not df.loc[index, column]:
            df.loc[index, column] = (
                df.loc[-1, "avg_person"]
                + df.loc[index, "person_diff"]
                + df.loc[-2, column]
            )
```
